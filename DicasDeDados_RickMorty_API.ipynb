```
import requests
import json
from pyspark.sql.functions import col, explode_outer
```

```
lista_ids = [ids for ids in range(0,100)]
```

```
url = f'https://rickandmortyapi.com/api/character/{lista_ids}'
```

```
dados = requests.get(url).json()
```

```
df = spark.createDataFrame(dados)
```

```
df2 = (
  df.withColumn('episode', explode_outer(col('episode')))
)
```

```
df2.select(
 'created',
 'episode',
 'gender',
 'id',
 'image',
 'name',
 'species',
 'status',
 'type',
 'url',
  col('location.name').alias('location_name'), col('location.url').alias('location_url'),
  col('origin.name').alias('origin_name'), col('origin.url').alias('origin_url'),
).display()
```
